```python
import pandas as pd
```

```python
 # Create a reference the CSV file desired
csv_path = "Resources/books2.csv"

# Read the CSV into a Pandas DataFrame
books2_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
books2_df.head()
```

```python
#dropped unwanted columns
Drop_column_bookdf = books2_df.drop(['bookID','language_code','Unnamed: 12'], axis=1)
#reorder columns
reorder_col_bookdf = Drop_column_bookdf[['isbn13', 'title', 'authors', '  num_pages','publisher','publication_date', 'isbn', 'average_rating', 'ratings_count', 'text_reviews_count',]]
#rename columns
rename_bookdf = reorder_col_bookdf.rename(columns={'isbn13': 'Book ID (Isbn-13)','title': 'Title', 'authors': 'Authors', 
                                                    '  num_pages': 'Number of Pages','publisher':'Publisher','publication_date':'Publication Date',
                                                    'isbn':'ISBN','average_rating':'Average Rating','ratings_count':'Ratings Count',
                                                    'text_reviews_count': 'Text Review Count'})
#new index
rename_booksdf = rename_bookdf.set_index('Book ID (Isbn-13)')
rename_booksdf.head()
```

```python
#set all review data into dataframe
book_review_df = rename_booksdf[['Average Rating','Ratings Count','Text Review Count']]
book_review_df.head()
```

```python
#set all book info into dataframe
book_info_df = rename_booksdf[['Title','Authors','Number of Pages','Publisher','Publication Date', 'ISBN']]
book_info_df.head() 
```

```python
df = rename_booksdf.loc[(rename_booksdf['Average Rating']>= '4.5') & (rename_booksdf['Ratings Count']>=100000), :]
df
```

```python
#export 2nd bookdf to csv
book_info_df.to_csv('book_info.csv')

#export cleaned df to csv (main csv file we'll be working with)
rename_booksdf.to_csv('book_df.csv')

#export ratings into csv
book_review_df.to_csv('book_ratings.csv')
```
