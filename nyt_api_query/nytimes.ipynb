```python
import json
import time
import requests
import pandas as pd
from config import api_key
from pprint import pprint

url = "https://api.nytimes.com/svc/books/v3/lists/best-sellers/history.json"
#isbn = "9780399169274"
```

```python
book_list_df = pd.read_csv('../book_df.csv')
print(book_list_df)
```

```python
most_rated_df = book_list_df.sort_values('Ratings Count', ascending=False)
most_rated_df
```

```python
#index = book_list_df['Book ID (Isbn-13)'].index
#index
ISBN_list = most_rated_df['Book ID (Isbn-13)'].to_list()
ISBN_list
```

```python
#response = requests.get(f"https://api.nytimes.com/svc/books/v3/lists/best-sellers/history.json?api-key={api_key}&isbn={ISBN_list[0]}").json()
#pprint(response)
```

```python
#isbn_list = []
response_list = []

#for isbn in isbn_list:
#    response = requests.get(f"https://api.nytimes.com/svc/books/v3/lists/best-sellers/history.json?api-key={api_key}&isbn={isbn}").json()
#    response_list.append(response)   
#    pprint(response) 

#pprint(response_list)

for i in range(500):
    response = requests.get(f"https://api.nytimes.com/svc/books/v3/lists/best-sellers/history.json?api-key={api_key}&isbn={ISBN_list[i]}").json()
    response_list.append(response)
    #pprint(response)
    time.sleep(6)
```

```python
pprint(response_list)
```

```python
list_of_isbn = []
title_list = []
author_list = []
publisher_list = []
description_list = []
nyt_link_list = []
bestseller_date_list = []
recent_rank_list = []
weeks_on_list = []

for row in range(len(response_list)):
    print(row)

    if response_list[row]['num_results'] == 0:
        print('Empty')
    else:      
        list_of_isbn.append(ISBN_list[row])

        try:
            title_list.append(response_list[row]['results'][0]['title'])
        except: 
            title_list.append('null')

        try:
            author_list.append(response_list[row]['results'][0]['author'])
        except:
            author_list.append('null')

        try:
            publisher_list.append(response_list[row]['results'][0]['publisher'])
        except:
            publisher_list.append('null')

        try:
            description_list.append(response_list[row]['results'][0]['description'])
        except:
            description_list.append('null')

        try:
            nyt_link_list.append(response_list[row]['results'][0]['reviews'][0]['sunday_review_link'])
        except:
            nyt_link_list.append('null')

        try:
            bestseller_date_list.append(response_list[row]['results'][0]['ranks_history'][0]['bestsellers_date'])
        except:
            bestseller_date_list.append('null')

        try:
            recent_rank_list.append(response_list[row]['results'][0]['ranks_history'][0]['rank'])
        except:
            recent_rank_list.append('null')

        try:
            weeks_on_list.append(response_list[row]['results'][0]['ranks_history'][0]['weeks_on_list'])
        except:
            weeks_on_list.append('null')
        
```

```python
len(list_of_isbn)
```

```python
main_table_df = pd.DataFrame({
    'ISBN-13': list_of_isbn,
    'Title': title_list,
    'Author': author_list,
    'Publisher': publisher_list,
    'Description': description_list,
    'NYT Review Link': nyt_link_list,
    'Most Recent Bestseller Date': bestseller_date_list,
    'Most Recent Rank': recent_rank_list,
    'Number of Weeks on List' : weeks_on_list,

})
main_table_df
```

```python
main_table_df.to_csv('main_table.csv')
```
